```python
# ------------------------------------------------------
# Program to demonstrate Off-by-One Error in Python
# ------------------------------------------------------

# Step 1: Define a list of 5 numbers
numbers = [10, 20, 30, 40, 50]

# ------------------------------------------------------
# Example 1: Off-by-One Error
# ------------------------------------------------------
print("âŒ Off-by-One Error Example:")

try:
    # The loop uses range(len(numbers) + 1)
    # len(numbers) = 5, so range(6) â†’ [0,1,2,3,4,5]
    for i in range(len(numbers) + 1):
        # When i = 0 â†’ numbers[0] â†’ 10 (OK)
        # When i = 4 â†’ numbers[4] â†’ 50 (OK)
        # When i = 5 â†’ numbers[5] â†’ ERROR (out of bounds)
        print(numbers[i])
except IndexError as e:
    # This block runs when the above code tries to access numbers[5]
    print("Error occurred:", e)

# ------------------------------------------------------
# Example 2: Corrected Version
# ------------------------------------------------------
print("\nâœ… Corrected Version:")

# Use range(len(numbers)) instead of len(numbers)+1
# range(5) â†’ [0,1,2,3,4] â†’ all valid indices
for i in range(len(numbers)):
    # Access each element safely using valid indices
    print(numbers[i])

# ------------------------------------------------------
# Example 3: Pythonic Way (No Risk of Off-by-One)
# ------------------------------------------------------
print("\nğŸ”¥ Pythonic Way (Better):")

# Directly loop through the list without using indices
for num in numbers:
    # Each element is accessed automatically
    print(num)
```

```python
# ------------------------------------------------------
# Program to demonstrate List Slicing in Python
# ------------------------------------------------------

# Step 1: Define a list of 5 numbers
numbers = [10, 20, 30, 40, 50]

# ------------------------------------------------------
# Example 1: Slice the entire list
# ------------------------------------------------------
print("ğŸ”¹ Full List Slice:")
# numbers[:] means take all elements from start to end
print(numbers[:])  

# ------------------------------------------------------
# Example 2: Slice the first 3 elements
# ------------------------------------------------------
print("\nğŸ”¹ First 3 Elements:")
# numbers[0:3] â†’ elements at index 0, 1, 2
print(numbers[0:3])  

# ------------------------------------------------------
# Example 3: Slice from index 2 to the end
# ------------------------------------------------------
print("\nğŸ”¹ From Index 2 to End:")
# numbers[2:] â†’ elements at index 2, 3, 4
print(numbers[2:])  

# ------------------------------------------------------
# Example 4: Slice with a Step
# ------------------------------------------------------
print("\nğŸ”¹ Every Second Element:")
# numbers[::2] â†’ start to end, step size 2
print(numbers[::2])  

# ------------------------------------------------------
# Example 5: Reverse the List
# ------------------------------------------------------
print("\nğŸ”¹ Reverse the List:")
# numbers[::-1] â†’ start to end backwards
print(numbers[::-1])  
```

```python
# ------------------------------------------------------
# Program to demonstrate Looping Through a List in Python
# ------------------------------------------------------

# Step 1: Define a list of 5 numbers
numbers = [10, 20, 30, 40, 50]

# ------------------------------------------------------
# Example 1: Using a for loop with range() and indices
# ------------------------------------------------------
print("ğŸ”¹ Loop using range() and indices:")

# range(len(numbers)) â†’ range(5) â†’ indices 0 to 4
for i in range(len(numbers)):
    # Access each element using its index
    print("Index:", i, " Value:", numbers[i])

# ------------------------------------------------------
# Example 2: Directly looping over elements (Pythonic way)
# ------------------------------------------------------
print("\nğŸ”¹ Loop directly over elements:")

# Each element is assigned to 'num' automatically
for num in numbers:
    print("Value:", num)

# ------------------------------------------------------
# Example 3: Using while loop with manual index
# ------------------------------------------------------
print("\nğŸ”¹ Loop using while:")

# Initialize index variable
i = 0
# Continue until i is less than length of the list
while i < len(numbers):
    print("Index:", i, " Value:", numbers[i])
    i += 1  # Increment index each time

# ------------------------------------------------------
# Example 4: Using enumerate() for index and value
# ------------------------------------------------------
print("\nğŸ”¹ Loop using enumerate():")

# enumerate(numbers) â†’ pairs of (index, value)
for index, value in enumerate(numbers):
    print("Index:", index, " Value:", value)
```

```python
# ------------------------------------------------------
# Program: Looping through a Slice, Copying a List, and Understanding Tuples
# ------------------------------------------------------

# Step 1: Define a list
numbers = [10, 20, 30, 40, 50, 60, 70]

# ------------------------------------------------------
# Part 1: Looping through a Slice of a List
# ------------------------------------------------------
print("ğŸ”¹ Looping through a slice (first 4 elements):")

# Slice syntax numbers[start:end]
# This takes elements from index 0 up to (but not including) index 4
for num in numbers[0:4]:
    print("Value:", num)

# ------------------------------------------------------
# Part 2: Copying a List
# ------------------------------------------------------
print("\nğŸ”¹ Copying a list:")

# Method 1: Using slicing
copy1 = numbers[:]  
print("Copy using slicing:", copy1)

# Method 2: Using list() constructor
copy2 = list(numbers)  
print("Copy using list() constructor:", copy2)

# Method 3: Using copy() method
copy3 = numbers.copy()  
print("Copy using copy() method:", copy3)

# ------------------------------------------------------
# Part 3: Understanding Tuples
# ------------------------------------------------------
print("\nğŸ”¹ Tuples demonstration:")

# Define a tuple with mixed data types
student = ("Anirudh", 21, "Computer Science")

# Print the tuple
print("Student Tuple:", student)

# Accessing elements by index
print("Name:", student[0])
print("Age:", student[1])
print("Department:", student[2])

# Looping through a tuple
print("\nLooping through tuple elements:")
for item in student:
    print(item)

# Tuples are immutable, so the following line would cause an error:
# student[1] = 22  # Uncommenting this will raise TypeError
```

```python
# ------------------------------------------------------
# Program: Looping Through a Tuple in Python
# ------------------------------------------------------

# Step 1: Define a tuple
fruits = ("Apple", "Banana", "Cherry", "Mango")

# ------------------------------------------------------
# Example 1: Loop directly through tuple elements
# ------------------------------------------------------
print("ğŸ”¹ Loop directly through elements:")
for fruit in fruits:
    # Each element is assigned to 'fruit' automatically
    print("Fruit:", fruit)

# ------------------------------------------------------
# Example 2: Loop using range() and indices
# ------------------------------------------------------
print("\nğŸ”¹ Loop using indices:")
for i in range(len(fruits)):
    # Access tuple element by index
    print("Index:", i, " Fruit:", fruits[i])

# ------------------------------------------------------
# Example 3: Loop using enumerate() to get index + value
# ------------------------------------------------------
print("\nğŸ”¹ Loop using enumerate():")
for index, fruit in enumerate(fruits):
    # enumerate gives both index and value
    print("Index:", index, " Fruit:", fruit)

# ------------------------------------------------------
# Example 4: While loop
# ------------------------------------------------------
print("\nğŸ”¹ Loop using while:")
i = 0
while i < len(fruits):
    print("Index:", i, " Fruit:", fruits[i])
    i += 1  # Move to the next index
```
